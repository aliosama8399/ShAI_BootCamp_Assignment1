```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv('/content/Salaries.csv')
df.head()
```

```python
print(f'The number of missing values in each column are:\n{df.isnull().sum()}')
```

```python
# Descriptive Statistics
print(f'The mean salary is ${df["TotalPayBenefits"].mean():,.2f}.')
print(f'The median salary is ${df["TotalPayBenefits"].median():,.2f}.')
print(f'The minimum salary is ${df["TotalPayBenefits"].min():,.2f}.')
print(f'The maximum salary is ${df["TotalPayBenefits"].max():,.2f}.')
print(f'The range of salaries is ${df["TotalPayBenefits"].max() - df["TotalPayBenefits"].min():,.2f}.')
print(f'The standard deviation of salaries is ${df["TotalPayBenefits"].std():,.2f}.')
```

```python
# df.dropna(inplace=True)
df.drop(columns=["Notes","Status"],inplace=True)
```

```python
df['BasePay'].fillna(df['BasePay'].mean(), inplace=True)
df['Benefits'].fillna(df['Benefits'].mean(), inplace=True)
df['OvertimePay'].fillna(df['OvertimePay'].mean(), inplace=True)
df['OtherPay'].fillna(df['OtherPay'].mean(), inplace=True)
```

```python
print(f'The number of missing values in each column are:\n{df.isnull().sum()}')
```

```python
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Basic Data Visualization
df['JobTitle'].value_counts().head(10).plot(kind='barh', figsize=(8, 6))
plt.title('Top 10 Job Titles')
plt.xlabel('Number of Employees')
plt.ylabel('Job Title')
plt.show()

# Grouped Analysis
grouped = df.groupby('Year')['TotalPayBenefits'].mean()
print(grouped)

# Simple Correlation Analysis
df.plot(kind='scatter', x='BasePay', y='TotalPayBenefits', figsize=(8, 6))
plt.title('Base Pay vs. Total Pay')
plt.xlabel('Base Pay')
plt.ylabel('Total Pay')
plt.show()

# Summary of Insights
print('Based on our analysis, we can conclude that the average salary has increased over the years. We also found that there is a strong correlation between base pay and total pay. Finally, we identified the top 10 job titles in the dataset.')
```
